# Prediction using Supervised ML
Simple Linear Regression

```python
#importing libraries
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt  
%matplotlib inline
from sklearn.model_selection import train_test_split  
from sklearn.linear_model import LinearRegression  
```

```python
#Creating a dataframe from the csv
df = pd.read_csv("http://bit.ly/w-data")
```

```python
df.head()
```

```python
df.describe()
```

```python
#Plotting the distribution of scores
fig = plt.figure(figsize = (15,10))
plt.scatter(df['Hours'], df['Scores'], color = 'Red')
plt.xlabel('Hours Studied')
plt.ylabel('Percentage Score')
plt.show()
```

```python
#dividing the data as attributes and target variable
X = df.iloc[:, :-1].values  
y = df.iloc[:, 1].values  
#splitting the data into train set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y , test_size = 0.2, random_state = 0)
```

```python
#model building and fitting the model with the train data
reg = LinearRegression()
reg.fit(X_train, y_train)
```

```python
# Plotting the regression line
line = reg.coef_* X + reg.intercept_
#plotting the test data
fig = plt.figure(figsize = (15,10))
plt.scatter(X, y, color = 'Red')
plt.plot(X, line)
plt.xlabel('Hours Studied')
plt.ylabel('Percentage Score')
plt.show()
```

```python
#making prediction
pred = reg.predict(X_test)
print(pred)
```

```python
#Comparing the predicted value with the actual value
op = pd.DataFrame({'Actual': y_test, 'Predicted': pred})  
op 
```
